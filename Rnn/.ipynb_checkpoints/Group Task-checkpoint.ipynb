```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, CuDNNLSTM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
import seaborn as sns
```

```python
train_set = pd.read_csv('train.csv',parse_dates=['Date'], low_memory = False)
final_set = pd.read_csv('test.csv', parse_dates=['Date'], low_memory = False)
```

```python
store_dataset = pd.read_csv('store.csv')
store_dataset
```

```python
merged_dataset = pd.merge(train_set, store_dataset, on = "Store", how = "left" )
merged_dataset = merged_dataset.drop(["Customers"], axis=1)
#merged_dataset.loc[merged_dataset['Store']==1]
merged_dataset
```

```python
merged_dataset_final = pd.merge(final_set, store_dataset, on = "Store", how = "left" )
merged_dataset_final = merged_dataset_final.drop(["Id"], axis=1)
#merged_dataset.loc[merged_dataset['CompetitionDistance'].isnull()]
merged_dataset_final
```

```python
merged_dataset_final.isnull().sum()
```

```python
CompetitionDistance = 'CompetitionDistance'
Promo2SinceWeek ='Promo2SinceWeek' 
CompetitionOpenSinceMonth = 'CompetitionOpenSinceMonth'
CompetitionOpenSinceYear = 'CompetitionOpenSinceYear'
Promo2SinceYear = 'Promo2SinceYear'
PromoInterval = 'PromoInterval'
Open = 'Open'
```

```python
if pd.api.types.is_numeric_dtype(merged_dataset[PromoInterval]):
    # Fill NaN values with the mean for numeric columns
    merged_dataset[PromoInterval].fillna(merged_dataset[PromoInterval].mean(), inplace=True)
    merged_dataset_final[PromoInterval].fillna(merged_dataset_final[PromoInterval].mean(), inplace=True)
else:
    # For non-numeric columns, fill NaN values with the most frequent value (mode)
    merged_dataset[PromoInterval].fillna(0, inplace=True)
    merged_dataset_final[PromoInterval].fillna(0, inplace=True)
```

```python
merged_dataset[CompetitionDistance].fillna(0, inplace=True)
merged_dataset[Promo2SinceWeek].fillna(0, inplace=True)
merged_dataset[CompetitionOpenSinceMonth].fillna(0, inplace=True)
merged_dataset[CompetitionOpenSinceYear].fillna(0, inplace=True)
merged_dataset[Promo2SinceYear].fillna(0, inplace=True)


merged_dataset_final[CompetitionDistance].fillna(0, inplace=True)
merged_dataset_final[Promo2SinceWeek].fillna(0, inplace=True)
merged_dataset_final[CompetitionOpenSinceMonth].fillna(0, inplace=True)
merged_dataset_final[CompetitionOpenSinceYear].fillna(0, inplace=True)
merged_dataset_final[Promo2SinceYear].fillna(0, inplace=True)
merged_dataset_final[Open].fillna(0, inplace=True)
```

```python
merged_dataset['StoreType']=lb.fit_transform(merged_dataset['StoreType'])
merged_dataset['Assortment']=lb.fit_transform(merged_dataset['Assortment'])
merged_dataset['StateHoliday']=lb.fit_transform(merged_dataset['StateHoliday'])

merged_dataset_final['StoreType']=lb.fit_transform(merged_dataset_final['StoreType'])
merged_dataset_final['Assortment']=lb.fit_transform(merged_dataset_final['Assortment'])
merged_dataset_final['StateHoliday']=lb.fit_transform(merged_dataset_final['StateHoliday'])
```

```python
merged_dataset['PromoInterval'] = merged_dataset['PromoInterval'].astype(str)
merged_dataset['PromoInterval']=lb.fit_transform(merged_dataset['PromoInterval'])

merged_dataset_final['PromoInterval'] = merged_dataset_final['PromoInterval'].astype(str)
merged_dataset_final['PromoInterval']=lb.fit_transform(merged_dataset_final['PromoInterval'])
```

```python
train_set_scalar =merged_dataset.iloc[:, 3:4].values
```

```python
print(train_set_scalar)
```

```python
train_y_list=[]
for i in train_set_scalar:
    train_y_list.append(i[0])
   
```

```python
total_y_set=pd.DataFrame(train_y_list,columns=['Norm_sales'])
total_y_set=total_y_set.astype(np.float32)
```

```python
total_y_set
```

```python
training_limit=int(len(total_y_set)*0.8)
```

```python
train_y_set=total_y_set[:training_limit]
test_y_set=total_y_set[training_limit:]
```

```python
total_x_set = train_set.drop(["Sales","Date","Customers","StateHoliday"], axis=1)
total_x_set=total_x_set.astype(float)
final_set= final_set.drop(["Id","Date","StateHoliday"], axis=1)
final_set=final_set.astype(float)
```

```python
train_x_set=total_x_set[:training_limit]
test_x_set=total_x_set[training_limit:]
```

```python
train_x_set
```

```python
final_set
```

```python
train_x_set_reshaped = np.reshape(train_x_set, (train_x_set.shape[0], 1, train_x_set.shape[1]))
test_x_set_reshaped = np.reshape(test_x_set, (test_x_set.shape[0], 1, test_x_set.shape[1]))
final_set=np.reshape(final_set, (final_set.shape[0], 1, final_set.shape[1]))
```

```python
train_x_set_reshaped.shape[2]
```

```python
reg = Sequential()
```

```python
reg.add(LSTM(128, input_shape=(1, 5), activation="relu",return_sequences=True))
reg.add(Dropout(0.2))
```

```python
reg.add(LSTM(128, activation="relu", return_sequences=True))
reg.add(Dropout(0.2))
reg.add(LSTM(64, activation="relu"))
reg.add(Dropout(0.2))
reg.add(Dense(30, activation='relu'))
reg.add(Dropout(0.2))
reg.add(Dense(10, activation='linear'))
```

```python
opt=tf.keras.optimizers.Adam(lr=1e-3,decay=1e-5)
```

```python
reg.compile(optimizer=opt, loss='mean_squared_error',
            metrics=["accuracy"])
```

```python
reg.fit(train_x_set_reshaped,train_y_set,epochs=1,validation_data=(test_x_set_reshaped,test_y_set))
```

```python
predictions=reg.predict(final_set)
```
